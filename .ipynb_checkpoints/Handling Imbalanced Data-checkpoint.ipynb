### Sampling

```python
To bring balance in the data we have to oversample or undersample it so that the count of class labels in the target column
are equal
There are two types of sampling
1. Oversampling
2. Undersampling
```

### Techniques

```python
1. Random Oversampling
2. Random Undersampling
3. SMOTE (Synthetic Minority Oversampling Technique)
```

### Random OverSampling

```python
import pandas as pd
import numpy as np

from imblearn.over_sampling import RandomOverSampler
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df
```

```python
df["Outcome"].value_counts()
```

```python
268/500
```

```python
df["Outcome"].value_counts().plot(kind="barh")
```

```python
random_oversampling = RandomOverSampler()
```

```python
x = df.drop("Outcome", axis = 1)
```

```python
y = df["Outcome"]
```

```python
x_oversampled, y_oversampled =  random_oversampling.fit_resample(x,y)
```

```python
y_oversampled.value_counts()
```

```python
x_oversampled.sort_values("")
```

### Random Over Sampling With Sampling Strategy

```python
random_oversampler = RandomOverSampler(sampling_strategy=0.7)
x_resampled, y_resampled = random_oversampler.fit_resample(x,y)
```

```python
y_resampled.value_counts()
```

```python
350/500
```

```python
random_oversampler = RandomOverSampler(sampling_strategy=0.9)
x_resampled, y_resampled = random_oversampler.fit_resample(x,y)
```

```python
y_resampled.value_counts()
```

```python
450/500
```

### Random Undersampling

```python
class1 >> 4500
class2 >> 5000
```

```python
from imblearn.under_sampling import RandomUnderSampler

random_undersampler = RandomUnderSampler()
x_undersampled, y_undersampled = random_undersampler.fit_resample(x,y)
```

```python
y_undersampled.value_counts()
```

## 3. SMOTE

```python
SMOTE >> Synthetic Minority Over Sampling Technique
```

```python
from imblearn.over_sampling import SMOTE
```

```python
y.value_counts()
```

```python
smote = SMOTE(random_state=10, k_neighbors=5)

x_oversampled, y_oversampled = smote.fit_resample(x,y)

y_oversampled.value_counts()
```

```python
x_oversampled.sort_values("BloodPressure")
```

```python
smote = SMOTE(k_neighbors=5)

x_oversampled, y_oversampled = smote.fit_resample(x,y)

y_oversampled.value_counts()
```
